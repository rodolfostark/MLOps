````python
new_column_names = {
    'vehicleType':'vehicle_type',
    'dateCrawled':'date_crawled',
    'offerType':'offer_type',
    'powerPS':'power_ps',
    'fuelType':'fuel_type',
    'yearOfRegistration':'registration_year',
    'monthOfRegistration':'registration_month',
    'notRepairedDamage':'unrepaired_damage',
    'dateCreated':'ad_created',
    'nrOfPictures':'nr_of_pictures',
    'postalCode':'postal_code',
    'lastSeen':'last_seen'
}
autos = autos.rename(columns = new_column_names)
autos.head()
````

With the dataframe property ```python3 DataFrame.describe()``` we could saw the most important data, select and drop out unuseful data 

````python
columns_to_be_dropped = ['seller', 'offer_type', 'abtest']
autos.drop(columns = columns_to_be_dropped)
````

````python
autos['price'] = [float(value.replace('$','').replace(',','')) for value in autos.price]
autos['odometer'] = [float(value.replace('km','').replace(',','')) for value in autos.odometer]
new_named_column = {
    'odometer':'odometer_km'
}
autos = autos.rename(columns = new_named_column)
autos.head()
````

Explorando o dataset, percebe-se que os carros cujo ano de registro são mais factíveis são os que estão entre os anos de 1900 e 2016

Com base na contagem na quantidade das marcas que mais aparecem podemos criar um dicionário a iterar sobre ele

````python
brand_list = autos.brand.value_counts().head(20).index
brand_list
````

````python
dictionary_cars_brand = {}
dictinary_odometer_mean = {}
for brand in brand_list:
    dictionary_cars_brand[brand] = autos[autos['brand'] == brand]['price'].mean()
    dictinary_odometer_mean[brand] = autos[autos['brand'] == brand]['odometer_km'].mean()
````

````python
dictionary_cars_brand
````

````python
dictinary_odometer_mean
````

````python
mean_brand_price = pd.Series(dictionary_cars_brand)
mean_odometer_km = pd.Series(dictinary_odometer_mean)
````

````python
dataframe_brand = pd.DataFrame(mean_brand_price, columns = ['mean_price'])
dataframe_brand
````

````python
dataframe_brand['mean_odometer_km'] = mean_odometer_km
dataframe_brand
````
